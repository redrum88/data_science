```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import os
df = pd.read_csv("https://raw.githubusercontent.com/redrum88/data_science/main/data/shooting_1982-2023_cleaned.csv")
import folium

# Create a map centered at the mean of the latitude and longitude columns
m = folium.Map(location=[df['latitude'].mean(), df['longitude'].mean()], zoom_start=4)

# Iterate over each row in the dataframe and add a marker for each case
for i, row in df.iterrows():
    popup_text = f"<b>{row['name']}</b><br>" \
                 f"<table>" \
                 f"<tr><td>Description:</td><td>{row['description']}</td></tr>" \
                 f"<tr><td>Fatalities:</td><td>{row['fatalities']}</td></tr>"\
                 f"<tr><td>Injured:</td><td>{row['injured']}</td></tr>" \
                 f"<tr><td>Total Victims:</td><td>{row['total_victims']}</td></tr>" \
                 f"<tr><td colspan='3'><hr></td></tr>" \
                 f"<tr><td>Age:</td><td>{row['age_of_shooter']}</td>" \
                 f"<td>Type:</td><td>{row['type']}</td>" \
                 f"<td>Year:</td><td>{row['year']}</td></tr>" \
                 f"<tr><td>Gender:</td><td>{row['gender']}</td>" \
                 f"<td>Month:</td><td>{row['month_name']}</td>" \
                 f"<td>Day of Week:</td><td>{row['day_of_week']}</td></tr>" \
                 f"<tr><td colspan='3'><hr></td></tr>" \
                 f"<tr><td>Race:</td><td>{row['race']}</td>" \
                 f"<td>Location:</td><td>{row['location']}</td>" \
                 f"<td>Weapon Type:</td><td>{row['weapon_type']}</td></tr>" \
                 f"<tr><td>Previously had signs of mental health issues:</td><td>{row['prior_signs_mental_health_issues']}</td>" \
                 f"<td>Mental health details:</td><td>{row['mental_health_details']}</td>" \
                 f"<td>Weapon obtained legally:</td><td>{row['weapons_obtained_legally']}</td></tr>" \
                 f"</table>"

    folium.Marker(location=[row['latitude'], row['longitude']], popup=popup_text,
                  tooltip=f"{row['fatalities']} fatalities").add_to(m)

# Save map
m.save('index.html')

# Display the map
m
```
